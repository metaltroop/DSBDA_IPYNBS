```python
import seaborn as sb
import matplotlib as plt
db=sb.load_dataset('titanic')
db.head()
```

```python
sb.boxplot(x='sex', y='age', data=db)
```

```python
sb.boxplot(x='sex', y='age', data=db, hue='survived')
```

```python
db['survived'].value_counts()
```

```python
mean=db['age'].mean()
mean
```

```python
std=db['age'].std()
std
```

```python
columns_to_check = ['age', 'sex', 'survived']
sb.heatmap(db[columns_to_check].isnull(), cmap='viridis', cbar=False)
```

```python
ds=db.select_dtypes(include=['float64','int64'])
corr01=ds.corr()
sb.heatmap(corr01)
```

```python
sb.countplot(data=db, x="survived", hue="alive")
```
