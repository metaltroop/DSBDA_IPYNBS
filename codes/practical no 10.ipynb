```python
  from sklearn.datasets import load_iris
import pandas as pd

# Load Iris dataset using scikit-learn
iris = load_iris()

# Create DataFrame from the loaded dataset
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['class'] = iris.target_names[iris.target]

# Display the first few rows of the dataset
iris_df.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class']
iris_df = pd.read_csv(url, names=names)
```

```python
print("Features and their types:")
print(iris_df.dtypes)
```

```python
iris_df.hist(figsize=(10, 6))
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the figure size
plt.figure(figsize=(10, 6))

# Create a boxplot for each feature excluding the 'class' column
sns.boxplot(data=iris_df.drop(columns='class'))

# Identify and plot outliers
for i, feature in enumerate(iris_df.columns[:-1]):
    q1 = iris_df[feature].quantile(0.25)
    q3 = iris_df[feature].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    outliers = iris_df[(iris_df[feature] < lower_bound) | (iris_df[feature] > upper_bound)]
    if not outliers.empty:
        plt.scatter([i]*outliers.shape[0], outliers[feature], color='pink', alpha=0.6)

# Add labels and title
plt.title('Boxplot of Features with Outliers Highlighted in Pink')
plt.xlabel('Features')
plt.ylabel('Values')

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(12, 8))
for i, feature in enumerate(iris_df.columns[:-1]):
    plt.subplot(2, 2, i + 1)
    sns.boxplot(x='class', y=feature, data=iris_df)
    plt.title(f'{feature.capitalize()} Distribution by Class')
    plt.xlabel('Class')
    plt.ylabel(feature)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create subplots
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))

# Plot histograms for each feature
sns.histplot(iris_df['sepal_length'], kde=True, color='blue', ax=axes[0, 0])
axes[0, 0].set_title('Sepal Length')

sns.histplot(iris_df['sepal_width'], kde=True, color='green', ax=axes[0, 1])
axes[0, 1].set_title('Sepal Width')

sns.histplot(iris_df['petal_length'], kde=True, color='yellow', ax=axes[1, 0])
axes[1, 0].set_title('Petal Length')

sns.histplot(iris_df['petal_width'], kde=True, color='red', ax=axes[1, 1])
axes[1, 1].set_title('Petal Width')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```
